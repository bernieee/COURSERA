```
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```
prices = pd.read_csv('close_prices.csv')
prices
```

```
prices_data = np.array(prices.drop(['date'], axis=1))
prices_data
```

```
n = 10
pca = PCA(n_components=n)
pca.fit(prices_data)

print(pca.explained_variance_ratio_)
varience = 0

for i in range(n):
    varience += pca.explained_variance_ratio_[i]

    if varience >= 0.9:
        break

print(i)
```

```
first_component = pca.transform(prices_data)[:, 0]
```

```
indexes = pd.read_csv('djia_index.csv')
indexes
```

```
indexes_data = np.array(indexes['^DJI'])
indexes_data
```

```
correlation = np.corrcoef(first_component, indexes_data)
correlation
```

```
index = pd.DataFrame(prices_data).columns[np.argmax(pca.components_[0])]
```
