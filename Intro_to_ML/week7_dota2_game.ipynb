```
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
import time
import datetime
from random import randint
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

## Data

```
features = pd.read_csv('features.csv', index_col='match_id')

features.head()
```

```
y = features['radiant_win']
```

```
features = features.drop(features.columns[102:], axis=1)

features.head()
```

```
features.shape
```

```
features[features.columns[0]].count()
```

```
num_of_values = features.shape[0]

for column in features.columns:
    if features[column].count() < num_of_values:
        print(column, '=', features[column].count())
```

```
features.fillna(0, inplace=True)
```

```
X = features.values
y = y.values
```

```
X
```

## Cross-validation

```
folds = KFold(n_splits=5, shuffle=True)

for train_id, test_id in folds.split(X):
    X_train, X_test = X[train_id], X[test_id]
    y_train, y_test = y[train_id], y[test_id] 
```

```
X_train.shape
y
```

## Logistic Regression

#### Basic Logistic Regression

```
scaler = StandardScaler()

X_train_scaled, X_test_scaled = scaler.fit_transform(X_train), scaler.transform(X_test)
```

```
c_list = np.linspace(0.01, 0.1, num = 10, endpoint = False)
c_list
```

```
for c in c_list:
    start_time = datetime.datetime.now()

    clf = LogisticRegression(C=c)
    clf.fit(X_train_scaled, y_train)

    print('Time elapsed:', datetime.datetime.now() - start_time)

    y_pred = clf.predict_proba(X_test_scaled)[:, 1]

    score = roc_auc_score(y_test, y_pred)
    print(score, 'c =', c)
    print('-----------------------------------------------------')
```

#### Unique Heros

```
hero_list = np.concatenate((features['r1_hero'], features['r2_hero'], features['r3_hero'], features['r4_hero'], features['r5_hero'],
                           features['d1_hero'], features['d2_hero'], features['d3_hero'], features['d4_hero'], features['d5_hero']))

hero_list_unique = set(hero_list)
num_of_heros = max(hero_list_unique)
print(hero_list_unique)
```

#### Bag Of Words

```
X_pick = np.zeros((features.shape[0], num_of_heros))

for i,  match_id in enumerate(features.index):
    for p in range(5):
        X_pick[i, features.loc[match_id, 'r%d_hero' % (p + 1)] - 1] = 1
        X_pick[i, features.loc[match_id, 'd%d_hero' % (p + 1)] - 1] = -1
```

```
X_pick.shape
```

```
drop_features_bag = drop_features
```

```
drop_features_bag = drop_features_bag.join(pd.DataFrame(X_pick, index=drop_features_bag .index)) 
drop_features_bag
```

```
X_bag = drop_features_bag.values
y_bag = y
```

```
for train_id, test_id in folds.split(X_bag):
    X_train_bag, X_test_bag = X_bag[train_id], X_bag[test_id]
    y_train_bag, y_test_bag = y_bag[train_id], y_bag[test_id] 
```

```
X_train_bag_scaled, X_test_bag_scaled = scaler.fit_transform(X_train_bag), scaler.transform(X_test_bag)
```

```
c = 0.1
start_time = datetime.datetime.now()

clf = LogisticRegression(C=c)
clf.fit(X_train_bag_scaled, y_train_bag)

print('Time elapsed:', datetime.datetime.now() - start_time)

y_pred = clf.predict_proba(X_test_bag_scaled)[:, 1]

score = roc_auc_score(y_test_bag, y_pred)
print(score, 'c =', c)
print('max =', max(y_pred))
print('min =', min(y_pred))
```
