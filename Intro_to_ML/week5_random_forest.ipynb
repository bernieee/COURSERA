```
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import r2_score
```

```
data = pd.read_csv('abalone.csv')
data
```

```
data['Sex'] = data['Sex'].map(lambda x:1 if x == 'M' else(-1 if x == 'F' else 0))
data
```

```
features = np.array(data.drop(['Rings'], axis=1))
aim = np.array(data['Rings'])
```

```
folds = KFold(n_splits=5, shuffle=True, random_state=1)
```

```
best_k = 0

for k in range(1, 51):
    regressor = RandomForestRegressor(n_estimators=k)
    #score = r2_score(y_true, y_pred, cv)
    score = cross_val_score(regressor, features, aim, cv=folds, scoring='r2')
    quality = np.mean(score)

    if quality >= 0.52:
        best_k = k
        break

print(best_k)
```
